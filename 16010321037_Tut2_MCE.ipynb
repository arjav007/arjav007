```python
var('x')
var('n')
f(x) = x^2
assume(n,'integer')
f = piecewise([[[-1,1],x^2]])
L = 1
an=(1/L)*integrate((x^2)*cos(n*pi*x),x,-1,1)
a0=(1/L)*integrate(f,x,-1,1)
bn=(1/L)*integrate((x^2)*sin(n*pi*x),x,-1,1)
s =a0/2+sum(an*cos(n*pi*x/L)+bn*sin(n*pi*x/L),n,1,5)
show(an)
show(bn)
show(a0)
show(s)
```

```python
plot(f,-1,1,legend_label="x^2") + plot(s,-1,1,color = "red",legend_label="fourier series")
```

```python
g = f.fourier_series_partial_sum(10)
show(g)
```
